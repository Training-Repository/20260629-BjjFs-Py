```python
val = 2

match val:
    case 1:
        print("One")
    case 2:
        print("Two")
    case 3:
        print("Three")
    case 4:
        print("Four")
    case 5:
        print("Five")
    case _:
        print("Outside the range")
```

```python
# match-case — structural routing of a loan request (clearer than nested if/elif)
def route(request):
    match request:
        case {"product": "home", "amount": amt} if amt > 5_000_000:
            return "Home loan (senior approval)"
        case {"product": "home"}:
            return "Home loan (standard)"
        case {"product": "personal", "amount": amt}:
            return f"Personal loan, amount {amt}"
        case _:
            return "Unrecognised request"

print(route({"product": "home", "amount": 8_000_000}))
print(route({"product": "home", "amount": 3_000_000}))
print(route({"product": "personal", "amount": 200000}))
print(route({"product": "gold"}))
```

```python
# A name is a reference to an object. Two names can point to the SAME object.
original = [1500, 320, 9800]
alias = original              # NOT a copy — same list object
alias.append(75)
print("original:", original)  # changed too!
print("same object?", original is alias)

copy = original[:]            # a real (shallow) copy
copy.append(0)
print("original:", original, "| copy:", copy)
print("same object?", original is copy)
```

### Complexity summary 
* **n** = number of elements
* **k** = slice length. 
* Marks: 
    * **\*** amortized · 
    * **†** average case (hash-based; degrades to O(n) worst under heavy collisions) · 
    * **—** not supported.

#### Time complexity by operation

| Operation | String | List | Tuple | Set | Frozenset | Dict | NamedTuple |
|---|---|---|---|---|---|---|---|
| Index by position `c[i]` | O(1) | O(1) | O(1) | — | — | — | O(1) |
| Lookup by key/field `c[k]` / `c.f` | — | — | — | — | — | O(1)† | O(1) |
| Slice `c[i:j]` | O(k) | O(k) | O(k) | — | — | — | O(k) |
| `len(c)` | O(1) | O(1) | O(1) | O(1) | O(1) | O(1) | O(1) |
| Membership `x in c` | O(n)‡ | O(n) | O(n) | O(1)† | O(1)† | O(1)† | O(n) |
| Append / add at end | — | O(1)\* | — | O(1)† | — | O(1)† | — |
| Insert at arbitrary position | — | O(n) | — | — | — | — | — |
| Delete by index / key / value | — | O(n) | — | O(1)† | — | O(1)† | — |
| Pop last | — | O(1) | — | O(1)† | — | O(1) | — |
| Pop first / by index | — | O(n) | — | — | — | — | — |
| Iterate all elements | O(n) | O(n) | O(n) | O(n) | O(n) | O(n) | O(n) |
| Copy | O(1)§ | O(n) | O(1)§ | O(n) | O(1)§ | O(n) | O(1)§ |

* ‡ Single-character membership is O(n); **substring** search (`"ab" in s`) is up to O(n·m) worst case.
* § Immutable types return the **same object** on copy (O(1)); building a fresh one by slicing/constructing is O(n).

#### Properties (for selection)

| Collection | Mutable | Ordered | Indexable | Duplicates | Hashable (usable as set member / dict key) |
|---|---|---|---|---|---|
| String | No | Yes | Yes | Yes | Yes |
| List | Yes | Yes | Yes | Yes | No |
| Tuple | No | Yes | Yes | Yes | Yes (if all items are) |
| Set | Yes | No | No | No | No |
| Frozenset | No | No | No | No | Yes |
| Dict | Yes | Yes (insertion) | By key | Keys unique | No |
| NamedTuple | No | Yes | Yes (+ by name) | Yes | Yes (if all items are) |

#### Selection takeaways

- **Fast membership / dedup → `set`/`frozenset`** (O(1)† vs O(n) for list/tuple).
- **Fast keyed lookup → `dict`** (O(1)†). Positional access → `list`/`tuple`.
- **List front-ops are O(n)** (`insert(0, …)`, `pop(0)`) — for a queue use `collections.deque` (O(1) both ends).
- **Building a string with `+=` in a loop is O(n²)** — use `"".join(parts)` (O(n)).
- **Immutable + hashable** (tuple, frozenset, namedtuple, string) is what lets a value serve as a dict key or set member; lists/sets/dicts can't.
