```python
# Booleans are a subtype of int: True == 1, False == 0
print(True + True, "<- yes, this is 2")
print(isinstance(True, int))

# Truthiness: empty / zero / None are falsy; everything else is truthy
for value in [0, 0.0, "", [], {}, None, 42, "SBI", [1]]:
    print(f"{str(value):>5}  -> {bool(value)}")
```
