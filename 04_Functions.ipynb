```python
def add(a:int, b:int)->int:
    print(f"{a = }")
    print(f"{b = }")
    sum = a + b
    # prod = a * b
    # lst = [sum, prod]
    return sum #, prod

print(add(1, 2))
print(add(1.1, 2.2))
print(add("1", "2"))
print(add("1", 2))
```

```python
## DOC STRINGS

# Multiplies 2 given objects
# a --> object type
# b --> object type
# Returns --> Object type
def mul(a:int, b:int)->int:
    """Multiplies 2 given objects
        a --> object type
        b --> object type
        Returns --> Object type
        """
    print(f"{a = }")
    print(f"{b = }")
    prod = a * b
    return prod

print(mul(1, 2))
print(mul(1.1, 2.2))
# print(mul("1", "2"))
print(mul("1", 2))
print(mul.__doc__)
```

```python
# Parameter handling

def add(a, b, c=0):
    print(f"{a = }")
    print(f"{b = }")
    print(f"{c = }")

    sum = a + b + c
    return sum

print(add(1, 2))            # Positional args
print(add(b=1, a=2))        # Named arg; Keyworded args
print(add(1, 2, 3))         # Oeverwrites the default argument 'c' with '3'
print(add(a=10))
```

```python
def add(a, b, c):
    print(f"{a = }")
    print(f"{b = }")
    print(f"{c = }")

    sum = a + b + c
    return sum

print(add(1, c=2, b=3))
```

```python
def separator(s1, occurance=10):
    print(s1 * occurance)

print(add(1, 2, 3))
separator('=', 80)
print(add(10, 20, 30))
```
