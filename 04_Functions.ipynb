```python
## DOC STRINGS

# Multiplies 2 given objects
# a --> object type
# b --> object type
# Returns --> Object type
def mul(a:int, b:int)->int:
    """Multiplies 2 given objects
        a --> object type
        b --> object type
        Returns --> Object type
        """
    print(f"{a = }")
    print(f"{b = }")
    prod = a * b
    return prod

print(mul(1, 2))
print(mul(1.1, 2.2))
# print(mul("1", "2"))
print(mul("1", 2))
print(mul.__doc__)
```

```python
def add(a, b, *c):          # Variable Arg list
    print(f"{a = }")
    print(f"{b = }")
    print(f"{c = }")

    sum = a + b
    for val in c:
        sum += val
    return sum

print(add(1, 2, 3, 4, 5, 6))
print(add(1, 2))
print(add(1, 2, *range(10)))
```

```python
# Function Overloading
from multipledispatch import dispatch

@dispatch(int, int)
def add(a, b):      # _add_int_int(a, b)
    return a+b

@dispatch(float, float)
def add(a, b):
    return a+b

print(add(1, 2))    # _add_int_int(1, 2)


@dispatch(int, int, int)
def add(a, b, c):   #_add_int_int_int(a, b, c)
    return a+b+c

print(add(1, 2, 3)) # _add_int_int_int(1, 2, 3)

print(add(1, 2))    # _add_int_int(1, 2)
print(add(1.1, 2.2))
```

```python
def func_1(num):    # Coroutine
    i = 0
    while i < num:
        i += 1
        yield i

def consumer1():
    gen = func_1(5)
    print(type(gen))
    print(next(gen))
    print(next(gen))
    print(next(gen))
    print(next(gen))
    print(next(gen))
    print(next(gen))

def consumer2():
    for val in func_1(5):
        print(val, end=" - ")
    print()

def consumer3():
    gen = func_1(5)
    try:
        while True:
            print(next(gen), end=" - ")
    except StopIteration:
        print("Done")    

consumer3()
```

```python
# Fibonacci Generator
def FibGen(n):
    '''Generates a sequence of 'n' fibonacci numbers'''
    a, b = 0, 1
    for i in range(n):
        yield a
        a, b = b, a+b

it = FibGen(10)
try:
    while(1):
        print(next(it))
except StopIteration:
    print("\nDone")
```

```python
class FibGen:
    def __init__(self, num):
        self.num = num

    def __iter__(self):
        self.a = 0
        self.b = 1
        self.count = 0
        return self
    
    def __next__(self):
        if self.count < self.num:
            x = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return x
        else:
            raise StopIteration


myFib = FibGen(10)
myIter = iter(myFib)

for x in myIter:
    print(x, end="-")
print()

for val in myFib:
    print(val, end="-")
print()

for val in myFib:
    print(val, end="-")
print()
```

```python
class Test:
    def __init__(self, data):
        self.data = data

    def Show(self):
        return f"Data -> {self.data}"
    
    def __str__(self):
        return f"My Data -> {self.data}"
    
    def __repr__(self):
        return f"[Data: {self.data} - {id(self.data)}]"

t1 = Test(100)
print(t1.data)
print(t1.Show())
print("-"*20)
print(t1)
print(str(t1))
print(t1.__str__())
print(t1.__repr__())
print(repr(t1))
```
